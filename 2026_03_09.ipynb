```python
import torch
from transformers import AutoTokenizer, AutoModelForMaskedLM
import torch.nn.functional as F

model_name = "klue/bert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForMaskedLM.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
inputs = tokenizer(text, return_tensors="pt").to(device)
mask_token_index = torch.where(inputs["input_ids"] == tokenizer.mask_token_id)[1]
```

```python
print(text)
print(mask_token_index)
```

```python
model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    predictions = outputs.logits
```

```python
mask_logits = predictions[0, mask_token_index]
probs = F.softmax(mask_logits, dim=-1)
```

```python
print(probs.sum())
print(probs.shape)
```

```python
top_values, top_indices = torch.topk(probs, k=10, dim=-1)
```

```python
for i in range(10):
    token_id = top_indices[0, i].item()
    token_prob = top_values[0, i].item()
    token_text = tokenizer.decode([token_id])

    rank = i + 1
    print(f"{rank}순위: {token_text} (ID: {token_id}, 확률: {token_prob:.4f})")
```

```python
def predict(text, k=10):
    inputs = tokenizer(text, return_tensors="pt").to(device)
    mask_token_index = torch.where(inputs["input_ids"] == tokenizer.mask_token_id)[1]

    model.eval()
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = outputs.logits


    mask_logits = predictions[0, mask_token_index]
    probs = F.softmax(mask_logits, dim=-1)


    print(probs.sum())
    print(probs.shape)

    top_values, top_indices = torch.topk(probs, k=k, dim=-1)
    return top_values, top_indices
```

```python
text = "저는 새로운 환경에서 항상 [MASK] 도전해왔습니다."
top_values, top_indices = predict(text)
```

```python
print(text)
for i in range(10):
    token_id = top_indices[0, i].item()
    token_prob = top_values[0, i].item()
    token_text = tokenizer.decode([token_id])

    rank = i + 1
    print(f"{rank}순위: {token_text} (ID: {token_id}, 확률: {token_prob:.4f})")
```

```python
model_name = "klue/roberta-large"
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = model.to('cuda')
```

```python
text = "냅둬라! 전과자들도 정치하고 약쟁이도 음악하고 방송하고 잘들사는 그런세상이다! "
inputs = tokenizer(text, return_tensor='pt', truncation=True, max_length=128, padding=True)
```

```python
def predict_sentiment(text):
    model.eval()
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        max_length=128,
        padding=True
    ).to(device)
    with torch.no_grad():
        outputs = model(**inputs)

    logits = outputs.logits
    probs = torch.softmax(logits, dim=-1)
    prediction = torch.argmax(probs, dim=-1).item()

    return prediction, probs[0]
```

```python
label, score = predict_sentiment(text)
print(f"label -> {label}, score -> {score}")
```
